### Car Data - missing values imputation

```python
data_car = data[data['class']=='car']
```

```python
data_car.shape
```

```python
data_car[data_car.isna().any(axis=1)].shape[0]
```

```python
data_car.isna().sum()
```

```python
mean_data = data_car.mean()
```

```python
#car_nan = data_car[data_car.isna().any(axis=1)]
```

```python
# replacing the nan values with their respective column means.
for i in range(0,18):
    #print(data_car.columns[i])
    i_nan = np.isnan(data_car.iloc[:,i])
    count = 0
    for a in i_nan:
        if a is True:
            #print(count)
            #print(data_car.iloc[count,i])
            data_car.iloc[count,i] = mean_data[i]
            #print(data_car.iloc[count,i])
        count = count+1
        
```

### SVM

```python
from sklearn import svm
from sklearn.model_selection import train_test_split
```

```python
y = vehicle_category
```

```python
X = vehicle_data_pca.iloc[:,:7]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, train_size=0.70, random_state=1)
```

```python
clf = svm.SVC(kernel = 'linear')
```

```python
clf.fit(X_train , y_train)
```

```python
clf.score(X_train , y_train)
```

```python
clf.score(X_test , y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'kernel':('linear', 'rbf'), 'C':[0.01,0.05,0.1,0.5,1]}
svc = svm.SVC(gamma="scale")
gs = GridSearchCV(svc, parameters, cv=50)
```

```python
gs.fit(X_train , y_train)
```

```python
gs.best_params_
```

```python
svc_best = svm.SVC(kernel = 'rbf', gamma="scale", C=1)
```

```python
svc_best.fit(X_train,y_train)
```

```python
svc_best.score(X_train,y_train)
```

```python
svc_best.score(X_test,y_test)
```

```python
y_pred = svc_best.predict(X_test)
```

```python
pd.crosstab(y_test,y_pred)
```
